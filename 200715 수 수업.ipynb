```python
from sklearn.metrics import roc_curve
```

```python
# fpr(거짓 양성률): 실제로는 해당사항이 없지만
# 해당 사항이 있다고 판정을 내리는 경우
# 암이 없는데 암이 있다는 판정 : ㅈ됨
fpr, tpr, thresholds = roc_curve(
    y_test, y_pred_prob
)
```

```python
print(fpr)
```

```python
print(tpr)
```

```python
print(thresholds)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr, tpr)
plt.title("ROC Curve")
plt.xlabel("False Positive rate")
plt.ylabel("True Positive rate")
plt.grid(True)
plt.show()
```

```python
def optimum_threshold(my_threshold):
    print(
        "Sensitivity: ",
          tpr[thresholds > my_threshold][-1]
    )
    print(
        "Specificity: ",
          1 - fpr[thresholds > my_threshold][-1]
    )
```

```python
print(tpr[thresholds > 0.5])
```

```python
# 여러값 나오는거 그냥 확인
print(tpr[thresholds > 0.4])
```

```python
# 파이썬은 배열을 거꾸로 순회할 수 있다.
# 맨 뒤의 배열을 가져옴
print(tpr[thresholds > 0.5][-1])
```

```python
# 0.3 모델
optimum_threshold(0.3)
```

```python
# 0.5 모델

# 0.3보단 0.5 모델을 쓰는게 낫다.
# 정확성(민감도)과 오판(특이성) 중 무엇을 택할것인가?
# 민감도와 특이성이 둘다 높으면 좋은 모델이지만
# 타협점을 찾을 필요가 있다.
optimum_threshold(0.5)
```

```python
optimum_threshold(0.01)
```

```python
optimum_threshold(0.0000001)
```
